```python
from src.measure_ir import *
import matplotlib.pyplot as plt
import numpy as np
from src.audio_device import AudioDevice
```

```python
fs = 48000
ad = AudioDevice(fs)
```

```python
ir_len = 128
dist = 7

irs = []
for i in range(8):
    error_ir, _ = measure_ir(ad)
    irs.append(align_ir_by_distance(error_ir, dist, ir_len, fs))
```

```python
plt.figure(figsize=(20, 10))
plt.plot(np.stack(irs).T)
plt.xlabel('Samples')
plt.ylabel('Amplitude')
plt.title('Comparison of Repeated IR Measurements After Alignment')
plt.show()
```
